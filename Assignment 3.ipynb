```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def optimal_thresholding(image):
    n = len(image)
    m = len(image[0])
    segmented_image = [[ 255 for _ in range(m)] for _ in range(n)]
    segmented_image[0][0] = segmented_image[0][m - 1] = segmented_image[n - 1][0] = segmented_image[n - 1][m - 1] = 0
    mean_bg = (image[0][0] + image[0][m - 1] + image[n - 1][0] + image[n - 1][m - 1]) / 4
    mean_obj = (sum([sum(image[i]) for i in range(n)]) - (mean_bg * 4))/ (n * m - 4)
    threshold = (mean_obj + mean_bg) / 2
    while (True):
        count_obj = count_bg = new_mean_obj = new_mean_bg = 0
        for i in range(n):
            for j in range(m):
                if (image[i][j] > threshold):
                    count_obj += 1
                    new_mean_obj += image[i][j]
                    segmented_image[i][j] = 255
                else:
                    count_bg += 1
                    new_mean_bg += image[i][j]
                    segmented_image[i][j] = 0
        new_mean_obj = new_mean_obj / count_obj
        new_mean_bg = new_mean_bg / count_bg
        if (new_mean_obj == mean_obj and new_mean_bg == mean_bg):
            break
        else:
            threshold = (new_mean_obj + new_mean_bg) / 2
            mean_obj = new_mean_obj
            mean_bg = new_mean_bg
    return segmented_image 
```

```python
image = Image.open("1.jpg").convert("L")
segmented_image = optimal_thresholding(np.array(image).tolist())
plt.subplots(1,2)
plt.subplot(121)
plt.imshow(image, cmap="gray")
plt.subplot(122)
plt.imshow(segmented_image, cmap="gray")
```
