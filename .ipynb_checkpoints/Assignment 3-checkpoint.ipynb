```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def CalculateHistogram(img):
    or_size = img.size
    width = or_size[0]
    height = or_size[1]
    levels = [0]*256
    for i in range(width):
        for j in range(height):
            shade = img.getpixel((i, j))
            levels[shade]= levels[shade] + 1
    return levels
```

```python
def CalculateCumulativeHistogram(levels):
    cum_histogram = []
    cur = 0
    i=0
    while i<len(levels):
        cur = cur + levels[i]
        cum_histogram.append(cur)
        i = i + 1
    return cum_histogram
```

```python
def CalculateEqualizedHistogram(cum_histogram):
    G = len(cum_histogram)
    S = cum_histogram[-1]
    eq_histogram = []
    hc0 = cum_histogram[0]
#     print('denom: ',(S-cum_histogram[0]))
    for p in range(len(cum_histogram)):
        q = int(round((G-1)*(cum_histogram[p]-hc0)/(S-hc0)))
        eq_histogram.append(q)
    return eq_histogram
```

```python
def CalculateEqualizedImage(img, eq_histogram):
    or_size = img.size
    width = or_size[0]
    height = or_size[1]
    eq_image = Image.new("L", (width, height))
    for i in range(width):
        for j in range(height):
            p = img.getpixel((i, j))
            q = eq_histogram[p]
            eq_image.putpixel((i, j), q)        
    return eq_image
```

```python
def HistogramEqualizationWrapper(img):
    histogram = CalculateHistogram(img)
    cumulative_histogram = CalculateCumulativeHistogram(histogram)
    equalized_histogram = CalculateEqualizedHistogram(cumulative_histogram)
    equalized_image = CalculateEqualizedImage(img, equalized_histogram)
    return equalized_image
```

```python
def SegmentOptimalThresholding(image):
    n = len(image)
    m = len(image[0])
    segmented_image = [[ 255 for _ in range(m)] for _ in range(n)]
    segmented_image[0][0] = segmented_image[0][m - 1] = segmented_image[n - 1][0] = segmented_image[n - 1][m - 1] = 0
    mean_bg = (image[0][0] + image[0][m - 1] + image[n - 1][0] + image[n - 1][m - 1]) / 4
    mean_obj = (sum([sum(image[i]) for i in range(n)]) - (mean_bg * 4))/ (n * m - 4)
    threshold = (mean_obj + mean_bg) / 2
    while (True):
        count_obj = count_bg = new_mean_obj = new_mean_bg = 0
        for i in range(n):
            for j in range(m):
                if (image[i][j] > threshold):
                    count_obj += 1
                    new_mean_obj += image[i][j]
                    segmented_image[i][j] = 255
                else:
                    count_bg += 1
                    new_mean_bg += image[i][j]
                    segmented_image[i][j] = 0
        new_mean_obj = new_mean_obj / count_obj
        new_mean_bg = new_mean_bg / count_bg
        if (new_mean_obj == mean_obj and new_mean_bg == mean_bg):
            break
        else:
            threshold = (new_mean_obj + new_mean_bg) / 2
            mean_obj = new_mean_obj
            mean_bg = new_mean_bg
    segmented_image = Image.fromarray(np.uint8(segmented_image))
    return segmented_image 
```

```python
im_num = 1
image = Image.open(str(im_num)+".jpg").convert("L")
seg_only = SegmentOptimalThresholding(np.array(image).tolist())
histo_only = HistogramEqualizationWrapper(image)
histo_seg = SegmentOptimalThresholding(np.array(histo_only).tolist())
plt.subplots(2,2,figsize=(16,12))
plt.subplot(2,2,1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.subplot(2,2,2)
plt.title("Segmenting the original image")
plt.imshow(seg_only, cmap='gray')
plt.subplot(2,2,3)
plt.title("Equalized Image")
plt.imshow(histo_only, cmap='gray')
plt.subplot(2,2,4)
plt.title("Segmented equalized Image")
plt.imshow(histo_seg, cmap='gray')

plt.savefig("out/"+str(im_num)+"_res.jpg")

seg_only.save("out/"+str(im_num)+"_seg.jpg")
histo_only.save("out/"+str(im_num)+"_eq.jpg")
histo_seg.save("out/"+str(im_num)+"_segEq.jpg")
```
